```python
import xarray as xr
from xcube.core.rectify import rectify_dataset
```

```python
src_ds = xr.open_dataset('inputdata/O_L2_0001_SNS_2017104102450_v1.0.nc')
src_ds
```

```python
src_ds.rtoa_8.plot.imshow(figsize=(16,16))
```

```python
dst_ds = rectify_dataset(src_ds, var_names='rtoa_8')
```

```python
dst_ds
```

```python
dst_ds.rtoa_8.plot.imshow(figsize=(16,16))
```

```python
import numpy as np
```

```python
bbox = np.logical_and(np.logical_and(src_ds.lon >= 1.2, src_ds.lon <= 1.5), np.logical_and(src_ds.lat > 52.45, src_ds.lat < 52.55))
bbox.plot.imshow(vmin=0, vmax=1)
```

```python
x = src_ds.x.where(bbox)
y = src_ds.y.where(bbox)
```

```python
x.plot.imshow()
```

```python
y.plot.imshow()
```

```python
x1 = int(x.min())
x2 = int(x.max())
y1 = int(y.min())
y2 = int(y.max())
```

```python
x1, y1, x2, y2
```

```python
var = src_ds.rtoa_8.isel(x=slice(x1, x2 + 1), y=slice(y1, y2 + 1))
```

```python
var.plot.imshow()
```
