```python
import xarray as xr
import numpy as np
from xcube.core.resampling import affine_transform_dataset
from xcube.core.resampling import resample_ndimage
from xcube.core.gridmapping import GridMapping
```

```python
def imshow(var):
    var.plot.imshow(vmin=0, vmax=4)
```

```python
res = 0.1
source_ds = xr.Dataset(
            data_vars=dict(
                refl=xr.DataArray(
                    np.array([
                        [0, 1, 0, 2, 0, 3, 0, 4],
                        [2, 0, 3, 0, 4, 0, 1, 0],
                        [0, 4, 0, 1, 0, 2, 0, 3],
                        [1, 0, 2, 0, 3, 0, 4, 0],
                        [0, 3, 0, 4, 0, 1, 0, 2],
                        [4, 0, 1, 0, 2, 0, 3, 0],
                    ], dtype=np.float64),
                    dims=('lat', 'lon'))
            ),
            coords=dict(
                lon=xr.DataArray(50.0 + res * np.arange(0, 8) + 0.5 * res, dims='lon'),
                lat=xr.DataArray(10.6 - res * np.arange(0, 6) - 0.5 * res, dims='lat')
            )
        )
source_ds               
```

```python
imshow(source_ds.refl)
```

```python
source_gm = GridMapping.from_dataset(source_ds)
source_gm.xy_bbox, source_gm.xy_res
```

```python
target_gm = GridMapping.regular((3, 3), (50, 10), 0.1, source_gm.crs)
# target_gm.ij_to_xy_transform, target_gm.xy_to_ij_transform
target_ds = affine_transform_dataset(source_ds, source_gm, target_gm)
imshow(target_ds.refl)
```
